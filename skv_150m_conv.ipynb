```python
def conv_manual(f, g):
    n = len(f)
    h = np.zeros(n)
    
    for j in range(n):
        # g_last_j_inverted = g[-j:][::-1]
        h[j] = np.sum(f[0:j]*g[0:j][::-1])
    
    h = np.array([np.sum(f[0:j]*g[0:j][::-1]) for j in range(n)])
    return h
```
